```python
df1.columns = df1.columns.str.strip()
```

```python
df1.columns
```

```python
df1[['day','month','year','Temperature','RH','Ws']] = df1[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df1.info()
```

```python
objects = [features for features in df1.columns if df1[features].dtype == 'O']
```

```python
for i in objects:
    if i!='Classes':
        df1[i] = df1[i].astype(float)
```
